```python
# Read all the data from the raw_data table into a pandas DataFrame

import pandas as pd
from sqlalchemy import create_engine

db_user = "postgres"
db_pass = ""
db_server = "Postgres"
db_port = 5432
db_name = "dataapi"

DB_URI = f"postgresql+psycopg2://{db_user}:{db_pass}@{db_server}:{db_port}/{db_name}"

engine = create_engine(DB_URI)

query = "SELECT * FROM raw_data"

with engine.connect() as connection:
    df = pd.read_sql(query, connection)

print(df.head())
```
